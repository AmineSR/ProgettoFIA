```python
from sklearn.model_selection import train_test_split
```

```python
trainset , testset = train_test_split(df, test_size=0.2 , random_state=0)
```

```python
trainset['Class'].value_counts()
```

```python
testset['Class'].value_counts()
```

```python
def encodage (df) :
    code = {'benign' : 0 ,
        'malignant' : 1
    
    }
    for col in df.select_dtypes('object') :
        df[col] = df[col].map(code)
    
    return df
```

```python
df.head()
```

```python
df.tail ()
```

```python
def imputation (df):
     return df.dropna(axis=0)
```

```python
def preprocessing (df):
    
    df=encodage (df)
    df=imputation (df)
    x=df.drop('Class', axis = 1)
    y=df['Class']
    return x,y
```

```python
xtrain , ytrain = preprocessing (trainset)
```

```python
xtest,ytest = preprocessing (testset)
```

# Modelisation 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score ,f1_score , confusion_matrix, classification_report
from sklearn.model_selection import learning_curve 
```

```python
dt_model = DecisionTreeClassifier(random_state=0)
```

```python
dt_model.fit(xtrain,ytrain)
```

```python
y_predicted=dt_model.predict(xtest)
```

```python
accuracy_score(ytest,y_predicted)
print(confusion_matrix (ytest ,y_predicted ))
print(classification_report(ytest ,y_predicted ))
```

```python
N , train_score, val_score = learning_curve(dt_model , xtrain,ytrain,cv=4 , scoring='f1', train_sizes=np.linspace(0.1, 1, 10))
plt.figure(figsize=(12,8))
plt.plot(N, train_score.mean(axis=1))
```

# Random Forrest

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.pipeline import make_pipeline 
from sklearn.feature_selection import SelectKBest , f_classif 
```

```python
r_forrest=make_pipeline(SelectKBest(f_classif, k=4) , RandomForestClassifier(random_state=0) )
```

```python
r_forrest.fit(xtrain,ytrain)
```

```python
r_forrest_predicted = r_forrest.predict (xtest)
```

```python
accuracy_score(ytest ,r_forrest_predicted )
```

```python
print(confusion_matrix (ytest , r_forrest_predicted))
print(classification_report(ytest , r_forrest_predicted))
```

# Optimizzazione del SVM

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
preprocessor = make_pipeline(PolynomialFeatures(2), SelectKBest(f_classif, k=4))
```

```python
SVM=make_pipeline(preprocessor, StandardScaler(), SVC(random_state=0))
```

```python
SVM
```

```python
hyper_params = {'svc__gamma':[1e-3, 1e-4],
                'svc__C':[1,10,100]}
               
```

```python
grid= GridSearchCV(SVM , hyper_params , scoring='recall', cv=5)
```

```python
grid.fit(xtrain,ytrain)
```

```python
y_g_predict = grid.predict(xtest)
```

```python
accuracy_score(ytest ,y_g_predict )
```

```python
print(confusion_matrix (ytest ,y_g_predict ))
print(classification_report(ytest , y_g_predict))
```

# Voting classifier

```python
from sklearn.ensemble import  VotingClassifier
```

```python
model1= LogisticRegression  ()
model2=DecisionTreeClassifier(random_state=0)
model3= KNeighborsClassifier (n_neighbors = 5)
```

```python
eclf1 = VotingClassifier(estimators=[ ('lr', model1), ('Tree', model2 ), ('KNN', model3)], voting='hard')
```

```python
eclf1.fit(xtrain,ytrain)
print(eclf1.score(xtest,ytest))
```

```python
eclf2 = VotingClassifier(estimators=[ ('lr', model1), ('Tree', model2 ), ('KNN', model3)], voting='soft')
```

```python
eclf2.fit(xtrain,ytrain)
print(eclf2.score(xtest,ytest))
```
